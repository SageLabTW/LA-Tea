```python
# 譜分群
import numpy as np
import matplotlib.pyplot as plt
np.set_printoptions(precision=2, suppress=True)

# G has three clusters 0~4, 5~9, 10~14
# edges inside the same cluster are weighted 1
# the edge {4,5} and {9,10} are weighted 0.1
L = np.zeros((15,15), dtype=float)
L[:5, :5] = -1
L[5:10, 5:10] = -1
L[10:, 10:] = -1
L[[4,5,9,10], [5,4,10,9]] = -0.1
L[np.arange(15), np.arange(15)] -= L.sum(axis=1)
vals,vecs = np.linalg.eigh(L)
plt.plot(np.arange(1,16), vals)
print(vals)
print(vecs[:,:3])
```

```python
# 樹圖的費德勒向量
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
np.set_printoptions(precision=2, suppress=True)
E = np.array([
    [0,1,0,3,0,5],
    [1,2,3,4,5,6]
]).T
g = nx.Graph()
g.add_edges_from(E)
# g = nx.random_labeled_tree(10)
L = nx.laplacian_matrix(g).toarray()
vals,vecs = np.linalg.eigh(L)
pos = vecs[:,1:3]
print("Eigenvalues:", vals)

k = 1
fig,ax = plt.subplots(1, 1, figsize=(8,6))
ax.axis("off")
for e in g.edges:
    ax.plot(*pos[e,:].T, color="black", alpha=0.3)
cplot = ax.scatter(*pos.T, c=vecs[:,k], vmin=-1, vmax=1, cmap="seismic", zorder=10)
ax.set_title(r"$\lambda_{%d} = %.4f$"%(k+1,vals[k]))

fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([0.85, 0.15, 0.05, 0.7])
fig.colorbar(cplot, cax=cbar_ax)
```
